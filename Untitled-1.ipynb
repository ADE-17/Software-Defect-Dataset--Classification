```python
import pandas as pd
```

```python
train_df = pd.read_csv('playground-series-s3e23/train.csv')
test_df = pd.read_csv('playground-series-s3e23/test.csv')
sub = pd.read_csv('playground-series-s3e23/sample_submission.csv')
```

```python
from supervised.automl import AutoML
```

```python
automl = AutoML(mode='Compete', 
                eval_metric='auc')
```

```python
automl.fit(train_df.drop('defects', axis=1),train_df.defects)
```

```python
preds = automl.predict_proba(test_df)
```

```python
df = pd.DataFrame()

df['id'] = test_df['id']
df['defects'] = preds[:,1]
```

```python
df.to_csv('sub2.csv', index=False)
```
